```python
def read_lines(file_path, encoding='utf-8'):
    lines = []
    with open(file_path, 'r', encoding=encoding) as file:
        for line in file:
            lines.append(line.strip())
    return lines

# Example usage
if __name__ == "__main__":
    # Path to the text file
    file_path = "/content/negative-words.txt"

    # Try reading with different encodings
    encodings_to_try = ['utf-8', 'latin-1', 'utf-16']
    for encoding in encodings_to_try:
        try:
            lines = read_lines(file_path, encoding)
            print("File read successfully with encoding:", encoding)
            break  # Stop trying encodings once successful
        except UnicodeDecodeError:
            print("Failed to read with encoding:", encoding)
```

```python
negative=set(lines)
```

```python
def process_files_negative(input_dir):
    for filename in os.listdir(input_dir):
        input_file = os.path.join(input_dir, filename)
        print(input_file)
        counter=0
        paragraph = read_paragraph(input_file)
        lowercase_words = tokenize_and_lowercase(paragraph)
        for i in lowercase_words:
          if i in negative:
            counter+=1
        df.loc[df["URL_ID"]==filename[:-4], "NEGATIVE SCORE"] = counter
```
