```python
import pandas as pd
df=pd.read_excel("/content/output_data_1.xlsx")
df.head()
```

```python
import os
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
nltk.download('punkt')
```

```python
def read_paragraph(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        return file.read()
```

```python
df.head()
```

```python
def process_files(input_dir):
    for filename in os.listdir(input_dir):
        input_file = os.path.join(input_dir, filename)
        print(filename,end=" ")
        paragraph = read_paragraph(input_file)
        sentences = sent_tokenize(paragraph)
        words = [word_tokenize(sentence) for sentence in sentences]
        # Calculate the number of sentences and words
        num_sentences = len(sentences)
        num_words = sum(len(sentence) for sentence in words)
        average_sentence_length = num_words / num_sentences
        df.loc[df["URL_ID"]==filename[:-4], "AVG SENTENCE LENGTH"] = average_sentence_length
```

```python
import syllables
from nltk.tokenize import word_tokenize
from collections import Counter
```

```python
def process_files(input_dir):
    for filename in os.listdir(input_dir):
        input_file = os.path.join(input_dir, filename)
        print(filename,end=" ")
        paragraph = read_paragraph(input_file)
        words = word_tokenize(paragraph)
        word_freq = Counter(words)
        mydict=dict()
        for word, freq in word_freq.items():
          mydict[word]=freq
        counter=0
        for i in mydict.values():
          if i==1:
            counter+=1
        total=0
        for i in mydict.values():
          total+=1
        df.loc[df["URL_ID"]==filename[:-4], "PERCENTAGE OF COMPLEX WORDS"] = counter*100/total
```

#<b>6. SYLLABLE PER WORD COUNT


We count the number of Syllables in each word of the text by counting the vowels present in each word. We also handle some exceptions like words ending with "es","ed" by not counting them as a syllable.

```python
import nltk
from nltk.tokenize import word_tokenize
import re

def count_syllables(word):
    word = re.sub(r'(es|ed)$', '', word, flags=re.IGNORECASE)
    num_vowels = len(re.findall(r'[aeiouAEIOU]+', word)) - len(re.findall(r'[aeiouAEIOU]e$', word, flags=re.IGNORECASE))
    return max(num_vowels, 1)
```

```python
def process_files(input_dir):
    for filename in os.listdir(input_dir):
        input_file = os.path.join(input_dir, filename)
        print(filename,end=" ")
        paragraph = read_paragraph(input_file)
        words = word_tokenize(paragraph)
        syllable_counts = {word: count_syllables(word) for word in words}
        mydict=dict()
        for word, syllables in syllable_counts.items():
          mydict[word]=syllables
        count=0
        for i in mydict.values():
          count+=1
        df.loc[df["URL_ID"]==filename[:-4], "SYLLABLE PER WORD"] = count
```

```python
process_files("/content/POST_STOPWORDS2")
```

#<b> 7.PERSONAL PRONOUNS


image.png

```python
import re

def count_personal_pronouns(text):
    # Define the pattern to match personal pronouns
    pattern = r'\b(I|we|my|ours|us)\b'
    # Find all matches in the text
    matches = re.findall(pattern, text, flags=re.IGNORECASE)
    # Exclude matches that are equal to 'us' when referring to the country
    matches = [match for match in matches if match.lower() != 'us']
    # Count the number of matches
    count = len(matches)
    return count

# Sample paragraph
paragraph = "I think we should go to my place. It's ours, after all. What do you think?"

# Count personal pronouns in the paragraph
pronoun_count = count_personal_pronouns(paragraph)

print("Personal Pronoun Count:", pronoun_count)
```

```python
def process_files(input_dir):
    for filename in os.listdir(input_dir):
        input_file = os.path.join(input_dir, filename)
        print(filename,end=" ")
        paragraph = read_paragraph(input_file)
        pronoun_count = count_personal_pronouns(paragraph)
        df.loc[df["URL_ID"]==filename[:-4], "PERSONAL PRONOUNS"] = pronoun_count

```

```python
process_files("/content/POST_STOPWORDS2")
```

#<b>8. AVG WORD LENGTH

```python
def average_word_length(paragraph):
    words = word_tokenize(paragraph)
    total_length = sum(len(word) for word in words)
    avg_length = total_length / len(words) if len(words) > 0 else 0
    return avg_length
```

```python
def process_files(input_dir):
    for filename in os.listdir(input_dir):
        input_file = os.path.join(input_dir, filename)
        print(filename,end=" ")
        paragraph = read_paragraph(input_file)
        df.loc[df["URL_ID"]==filename[:-4], "AVG WORD LENGTH"] = average_word_length(paragraph)
```
